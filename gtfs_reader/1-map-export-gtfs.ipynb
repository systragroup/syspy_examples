```python
qpath = r'../../quetzal/' # path to quetzal here
import sys
sys.path.insert(0, qpath)

data = r'inputs/'

# import class
from quetzal.io.gtfs_reader.feed_gtfsk import Feed
```

### Read GTFS

```python
feed = Feed(path=data + r'accra_trotro', dist_units='m')
feed.describe()
```

```python
feed.validate()
```

### Map GTFS

```python
feed.map_stops(
    feed.stops['stop_id']  # IDs of the stops to plot
)
```

The geometry of a route is the results of the shapely.linemerge operation of all its trips geometries.

```python
feed.map_routes(
    feed.routes['route_id'].head(50)  # IDs of the routes to plot
)
```

```python
feed.map_trips(
    feed.trips.groupby('route_id').first().trip_id.head(50)  # IDs of the trips to plot: one per route
)
```

```python
feed.trips.groupby('route_id')['trip_id'].count()
```

### Export as GIS

```python
feed.write_geojson(
    'outputs/accra/' # folder name in which will be written 3 geojson files: routes, stops and trips
)
```
